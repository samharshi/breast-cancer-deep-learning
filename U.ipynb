importing the data dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

data collection and processing (loading the data from the sklearn

```python
breast_cancer_dataset=sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
#loading the  data toa data frame
data_frame=pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
```

```python
#print the first 5 rows of the data frame
data_frame.head()
```

```python
#adding the target column to the data frame
data_frame['Label']=breast_cancer_dataset.target
# tf.random.set_seed(3) ensures reproducibility of random operations (e.g., weight initialization, data splitting) in TensorFlow.
```

seperating features and targets

```python
X=data_frame.drop(columns='Label',axis=1) #when ever want t drap a column we want to use axis=1,row axis=0
Y=data_frame['Label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

standadize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_train_std=scaler.fit_transform(X_train)
X_test_std=scaler.transform(X_test)
```

```python
print(X_train_std)
```

Building the neural network

```python
#importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#setting up the layers neural network
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(2,activation='sigmoid')
])
```

```python
#compiling the neural network
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

dog-[0 0 1 ]
cat-[0 1 0 ]
horse-[0 0 1 ]

```python
#training the neural network
history=model.fit(X_train_std,Y_train,validation_split=0.1,epochs=10) #how many times the model has to go through the data
```

here in the above loss function predicted what is the true value and what is the predicted value

visualizing

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data','validation data'],loc='lower right')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_accuracy'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data','validation data'],loc='upper right')
plt.show()
```

accuracy of the model and the test data

```python
losss,accuracy=model.evaluate(X_test_std,Y_test)
print(accuracy)
```
